```python
# Create a folder for storing consultation files
if not os.path.exists("consultations"):
    os.mkdir("consultations")
```

```python
consultations = Consultation.objects.all()
```

```python
# Create a zip file to store the consultations
with zipfile.ZipFile('consultations.zip', 'w') as zipf:
    for idx, consultation in enumerate(consultations):
        # Format each consultation
        consultation_text = format_consultation(consultation)
        
        # Create a text file for each consultation
        file_name = f"consultations/consultation_{idx + 1}.txt"
        with open(file_name, 'w') as file:
            file.write(consultation_text)
        
        # Add the text file to the zip
        zipf.write(file_name)
```

```python

# Clean up text files after zipping (optional)
for file_name in os.listdir("consultations"):
    os.remove(os.path.join("consultations", file_name))

os.rmdir("consultations")

print("Consultations have been exported to consultations.zip")
```

```python
import os
import zipfile
from django.http import HttpResponse
from django.core.files.storage import default_storage
from django.conf import settings
from io import BytesIO
```

```python
def generate_consultation_zip():
    # Creating an in-memory zip file
    in_memory = BytesIO()
    with zipfile.ZipFile(in_memory, 'w', zipfile.ZIP_DEFLATED) as zip_file:
        # Fetch all consultations
        consultations = Consultation.objects.all()
        for consultation in consultations:
            formated_consultation = format_consultation(consultation)
        # Create a file name based on the consultation id or name
        filename = f"consultation_{consultation.id}.txt"
        # Write the text content to the zip file
        zip_file.writestr(filename, formated_consultation)

    # Move the cursor of the in-memory file to the beginning
    in_memory.seek(0)
    
    return in_memory
```

```python
generate_consultation_zip()
```
