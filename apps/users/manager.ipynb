```python
# Loop through each row in the DataFrame and create User objects.
for index, row in df.iterrows():
    # Extract data from the DataFrame row.
    first_name = row['first_name']
    last_name = row['last_name']
    is_active = row['is_active']
    print(row['studentprofileimage__image'])
    image = None
    if isinstance(row['studentprofileimage__image'], str):
        image  = SimpleUploadedFile(row['studentprofileimage__image'], b"") 
    old_id = row['id']
    is_disabled = row['is_disabled_person']
    place_of_birth = row['place_of_birth']  # You can adjust this as needed.
    genre = row['genre']
    birth_date = datetime.strptime(str(row['born_on']), '%Y-%m-%d').date()
    email = f"{row['registration_id']}@gmail.com"  # Replace 'email' with the actual column name in your Excel file.
    if is_active:
        user = User.objects.create(
            first_name=first_name,
            last_name=last_name,
            old_id=old_id,
            is_disabled=is_disabled,
            place_of_birth=place_of_birth,
            birth_date=birth_date,
            genre=genre,
            email=email,
        )
        user.username = row['registration_id']
        user.save()
        profile = UserProfile.objects.create(user=user)
        profile.image=image
        profile.save()

print("Users created successfully.")
```

```python
# Loop through each row in the DataFrame and create User objects.
for index, row in df.iterrows():
    is_active = row['is_active']
    if is_active:
        student = Student.objects.filter(registration_id=row['registration_id']).first()
        
        student.specialty = row['specialty__name']
        student.save()
        print(student.specialty)


print("Users created successfully.")
```

```python
from apps.cameis.models import EducationYear


education_year = EducationYear.objects.first()
for student in students:
    education_class = EducationClass.objects.filter(option__option__name=student.specialty, option__path_cursus_level_specialty__path_cursus_level__level__value=student.level).first()
    if education_class:
        EnrolledStudent.objects.create(organization=organization,student=student,education_class=education_class,education_year=education_year)
    else:
        print(student,student.specialty, student.level)
```

```python
enrolled_students = EnrolledStudent.objects.all()
latest_year = EducationYear.objects.last()
```

```python
for enrolled_student in enrolled_students:
    next_level = enrolled_student.education_class.option.path_cursus_level_specialty.path_cursus_level.level.value + 1
    student_option_id = enrolled_student.education_class.option.option_id
    next_education_class = EducationClass.objects.filter(option__option_id=student_option_id, option__path_cursus_level_specialty__path_cursus_level__level__value=next_level).first()
    if next_education_class:
        print(enrolled_student.education_class, "next:", next_education_class)
        EnrolledStudent.objects.create(organization=organization,student=enrolled_student.student,education_class=next_education_class,education_year=latest_year)
```
