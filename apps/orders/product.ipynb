```python
import os
import django

os.environ.setdefault('DJANGO_SETTINGS_MODULE', 'focalflow.settings.dev')
os.environ["DJANGO_ALLOW_ASYNC_UNSAFE"] = "true"
django.setup()
```

```python
org = "92af06ea-0d1b-4e1c-ab39-82c331c70c27"
```

```python
from apps.orders import models as order_models
```

```python
items = order_models.Item.objects.filter(organization="f54f7c46-6891-4246-8124-4a1cf1d1ac06")
```

```python
product_list = []
for product in items:
    item = order_models.Item()
    item.organization_id = org
    item.name = product.name
    item.quanta = product.quanta
    item.purchase_price = product.purchase_price
    item.facturation_price = product.facturation_price
    item.quantity = 0
    item.alert_quantity = 0
    item.last_checked = product.last_checked
    item.last_maintainer = product.last_maintainer
    item.is_active = product.is_active
    item.is_service = product.is_service
    product_list.append(item)
```

```python
order_models.Item.objects.bulk_create(product_list)
```
