```python
import os
import django

os.environ.setdefault('DJANGO_SETTINGS_MODULE', 'focalflow.settings.dev')
os.environ["DJANGO_ALLOW_ASYNC_UNSAFE"] = "true"
django.setup()
```

```python
from apps.orders import models as order_models
from apps.cashflow import models as cashflow_models
```

```python
deposits = cashflow_models.Deposit.objects.all()
withdrawals = cashflow_models.Withdrawal.objects.all()
```

```python
transactions = []
for deposit in deposits:
    transactions.append(
        order_models.Transaction(
            created = deposit.created,
            modified = deposit.modified,
            organization=deposit.organization,
            organization_user=deposit.organization_user,
            transaction_broker=order_models.TransactionBroker.CASHIER,
            transaction_type=order_models.TransactionType.DEPOSIT,
            amount=deposit.amount,
            participant="Docta Shop",
            reason=deposit.reason,
        )
    )
    
for withdrawal in withdrawals:
    transactions.append(
        order_models.Transaction(
            created = withdrawal.created,
            modified = withdrawal.modified,
            organization=withdrawal.organization,
            organization_user=withdrawal.organization_user,
            transaction_broker=order_models.TransactionBroker.CASHIER,
            transaction_type=order_models.TransactionType.WITHDRAWAL,
            amount=withdrawal.amount,
            participant=withdrawal.recipient,
            reason=withdrawal.reason,
        )
    )
```

```python
payments = order_models.FacturationPayment.objects.all()
```

```python
payments2 = []
for payment in payments:
    payments2.append(
        order_models.FacturationPayment(
            created = payment.created,
            modified = payment.modified,
            facturation = payment.facturation,
            organization = payment.organization,
            organization_user = payment.facturation.organization_user,
            transaction_broker = order_models.TransactionBroker.CASHIER,
            amount = payment.amount
        )
    )
```

```python
created_transactions = order_models.Transaction.objects.bulk_create(transactions)
```

```python
created_payments = order_models.FacturationPayment.objects.bulk_create(payments2)
```

```python
for tran in created_transactions:
    print(tran.created)
```
