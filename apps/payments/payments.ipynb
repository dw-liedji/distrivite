## Storing users payments

```python
from payments.models import Payment, PaymentMethod
```

```python
MTNPayment = PaymentMethod.objects.filter(pk=1).first()
OrangePayment = PaymentMethod.objects.filter(pk=2).first()
```

```python
from accounts.models import User
user1 = User.objects.filter(pk=1).first()
```

```python
from subscriptions.models import Plan
```

```python
plan1 = Plan.objects.filter(pk=1).first()
```

```python
payment1 = Payment()
payment1.method = MTNPayment
payment1.reference = "95287e26-6b95-4930-a6a4-fe0c6d4ed200"
payment1.amount = "200"
payment1.phone = '+237672120240'
payment1.user = user1
payment1.generic_product_object = plan1
```
